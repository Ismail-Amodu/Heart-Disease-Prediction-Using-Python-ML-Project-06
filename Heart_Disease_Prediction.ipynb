```python
# Importing the dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

** DATA COLLECTION AND PROCESSING**

```python
# Loading the csv dataset into pandas dataframe
heart_data = pd.read_csv('/content/heart_disease.csv')
```

```python
# priinting the first five rows of the dataframe
heart_data.head()
```

```python
# priinting the last five rows of the dataframe
heart_data.tail()
```

```python
# getting the number of rows and column in the dataframe
heart_data.shape
```

```python
# Getting some info about the data
heart_data.info()
```

```python
# Checking for missing values
heart_data.isnull().sum()
```

```python
new_heart_data = heart_data.dropna(how='any')
```

```python
# Obtaining the statistical measures of the data
new_heart_data.describe()
```

DISTRIBUTION OF THE TARGET VARIABLEE

1.   Healthy heart --> 0
2.   Defective heart --> 1


```python
# Checking the distribution of the target variable to see if there is almost equal values
new_heart_data['HeartDisease'].value_counts()
```

**Handling the imbalance in the dataset**


1.   healthy_heart --> 0
2.   defective_heart --> 1

```python
# Separating the dataset for analysis
healthy_heart = new_heart_data[new_heart_data.HeartDisease == 0]
defective_heart = new_heart_data[new_heart_data.HeartDisease == 1]
```

```python
print(healthy_heart.shape)
```

```python
print(defective_heart.shape)
```

```python
# Comparing the values for both transactions
new_heart_data.groupby('HeartDisease').mean()
```

**UNDER-SAMPLING**

```python
# Bulding  a sample dataset containing similar distribution of healthy heart and defective heart

healthy_heart_sample = healthy_heart.sample(n=5735)
```

```python
# Concatenating two dataframes
new_dataset = pd.concat([healthy_heart_sample, defective_heart], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['HeartDisease'].value_counts()
```

```python
new_dataset.groupby('HeartDisease').mean()
```

**SPLITTING DATASET**

```python
# Splitting data into features and targets
X = new_dataset.drop(columns='HeartDisease', axis=1)
Y = new_dataset['HeartDisease']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting data into training and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=5)
print(X.shape, X_train.shape, X_test.shape)
```

**MODEL TRAINING: Logistic regression Model**

```python
#  Loading the logistic regression model in an object, then increasing the number of iteration from default (100) to 1000
#  This approach ensures convergence of the optimization for minimizing cost function
model = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
# Training the model with dataset
model.fit(X_train, Y_train)
```

**MODEL EVALUATION**

```python
# Accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy on training data : ', training_data_accuracy)
```

```python
# Accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy on test data : ', test_data_accuracy)
```
